```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data_a=pd.read_excel('../阶段一恒a数据.xlsx')
data_F=pd.read_excel('../阶段一恒F数据.xlsx')
```

```python
data_a.head()
```

```python
data_F.head()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']

plt.figure(figsize=(10, 6))
plt.plot(data_a['时间_s'], data_a['质量_kg'], label='恒a策略', color='blue', linewidth=2)
plt.plot(data_F['时间_s'], data_F['质量_kg'], label='恒F策略', color='red', linewidth=2)
plt.text(data_a['时间_s'].iloc[-1], data_a['质量_kg'].iloc[-1], f'{data_a["质量_kg"].iloc[-1]:.2f} kg', fontsize=12, color='blue')
plt.text(data_F['时间_s'].iloc[-1], data_F['质量_kg'].iloc[-1], f'{data_F["质量_kg"].iloc[-1]:.2f} kg', fontsize=12, color='red')
plt.xlabel('时间 (s)', fontsize=12)
plt.ylabel('质量 (kg)', fontsize=12)
plt.title('恒a与恒F策略下质量随时间变化对比', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
